```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x = np.random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(15))
print(x)

x = np.random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(3,5))
print(x)
```

# Shuffling array

```python
arr = np.array([1,2,3,4,5])
np.random.shuffle(arr) #The shuffle() method makes changes to the original array.
print(arr)
```

```python
arr = np.array([1,2,3,4,5])
print(np.random.permutation(arr)) #The permutation() method returns a re-arranged array (and leaves the original array un-changed).
```

```python
numbers = np.random.normal(loc=0, scale=5 , size=(10000))
sns.displot(numbers, kind='hist', kde=True)
plt.show()
```

```python
numbers = np.random.normal(loc=0, scale=5 , size=(2,2))
numbers
```

```python
x = np.random.binomial(n=10, p=0.5, size=1000)
sns.displot(x, kde=True)
plt.show()
```

```python
arr = np.random.poisson(lam=2, size=1000)
# print(arr)

sns.displot(arr, kind='kde')

plt.show()
```

```python
arr1 = np.random.normal(loc=50, scale=7, size=1000)
arr2 = np.random.poisson(lam=50, size=1000)

arr = np.vstack((arr1, arr2))
print(arr)

sns.displot(arr.T, kind='kde')
plt.show()
```
