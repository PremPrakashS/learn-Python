# Subsetting Data

```python
df_new.columns
```

```python
#subsetting columns
df_new = df_new[['title', 'description', 'publishTime', 'duration_seconds',
       'viewCount', 'likeCount', 'commentCount']]
```

```python
# df_new.head()
df_new.shape
```

```python
#subsetting rows
df_subset1 = df_new.loc[df_new['viewCount'] > 100_000_000]
```

```python
df_subset2 = df_new.query('viewCount > 100000000')
```

```python
# ~df_new['viewCount'].isna()

df_new = df_new.loc[~df_new['viewCount'].isna()]
```

# Casting dtypes

```python
df_new.dtypes
```

```python
df_new['viewCount'] = df_new['viewCount'].astype('int') # cannot convert to int if the columns contains infinate of null values
# df_new['likeCount'] = df_new['likeCount'].astype('int')
```

```python
df_new.dtypes
```

```python
df_new['publishTime'] = pd.to_datetime(df_new['publishTime'])
```

```python
df_new['likeCount'] = pd.to_numeric(df_new['likeCount'].astype('str'))
```

# Creating new Column

```python
df_new['like_view_ratio'] = df_new['likeCount'] / df_new['viewCount']
```

```python
df_new.head()
```

# Adding a row

```python
df_to_append = df_new.tail(1)
```
