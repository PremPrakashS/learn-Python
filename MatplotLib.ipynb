```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpts = np.array([0, 6]) # declaring two points and ploting them
ypts = np.array([1, 4])
# print(xpts, ypts)

plt.plot(xpts, ypts) # plot is used to draw points in a diagram, by default, 
# the function draws a line from point to point
# parameter 1 is an array containing the points on the x-axis.
# parameter 2 is an array containing the points on the y-axis.
plt.show()
```

```python
plt.plot([1,4,8] , [3,1,10]) #plot(<points on x-axis>, <points on y-axis>)
plt.show()
```

```python
plt.plot([1,4,8] , [3,1,10], 'o') # small 'o' to plot points
plt.show()
```

# Format strings # fmt marker|line|color

```python
ypoints = np.array([3, 8, 1, 10])
```

```python
plt.plot(ypoints, 'o:r')  # color controls the color of the line as well as the marker
plt.show()
```

```python
plt.plot(ypoints, 'x-.c') # marker|line style|color
plt.show()
```

# Marker Size

```python
plt.plot(ypoints,'*:r' , ms = 20) # to control the marker size
```

# Marker Color

```python
plt.plot(ypoints, '*:y', ms=20, mec='r' , mfc='b') # mec = marker edge color / mfc = marker face color 
# hexa decimal values can be used for this mec = '#4CAF50', mfc = '#4CAF50' / mec = 'hotpink', mfc = 'hotpink'
plt.show()
```
